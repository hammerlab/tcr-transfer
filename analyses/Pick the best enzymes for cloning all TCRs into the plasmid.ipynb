```python
from Bio import SeqIO
from Bio import Restriction

import pandas as pd
```

```python
nt_records = [rec for rec in SeqIO.parse("../data/10X-VDJ-summary.all_nt.fasta", "fasta")]
```

![pcDNA3.1 map](https://cl.ly/d5fe78699be0/Image%2525202019-08-06%252520at%25252012.26.00%252520PM.png)

```python
mcs_enzymes = [Restriction.NheI,
               Restriction.AflII,
               Restriction.HindIII,
               Restriction.KpnI,
               Restriction.BamHI,
               Restriction.EcoRI,
               Restriction.EcoRV,
               Restriction.NotI,
               Restriction.XhoI,
               Restriction.XbaI,
               Restriction.ApaI]
```

```python
r_counts = pd.DataFrame([[len(e.search(tcr.seq)) for e in mcs_enzymes] for tcr in nt_records],
                        columns=[str(e) for e in mcs_enzymes],
                        index=[tcr.name for tcr in nt_records])

r_counts
```
